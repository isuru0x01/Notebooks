Now, we looked at the data and checked the variables. The first step is to go deeper and understand the variables. Here, the questions have to be asked: "which variables are really numerical, and which of these are categorical? Is there any cardinal variable?".

For instance, the target variable, Outcome, seems as a numerical variable, but it is known that it is categorical since this column only includes 1 and 0(disease or no disease). These kinds of variables have to be considered categorical.

Similarly, if there is a categorical variable that includes the names of the patients in a dataset, there would be considered a cardinal dataset because a name cannot carry any information. This is not valid for our dataset because there are no such columns, but it had to be considered if the dataset had.

Let's write a function to differentiate the types of variables(grab_col_names). The values are open to comment. In this problem, I considered that if a numerical variable has less than 10 different values, it is actually categorical. If a categorical value has more than 20 different values, it is cardinal. Since our all variables are numerical, there will not be any cardinal variable.

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    """
    This function identifies and categorizes columns in a pandas DataFrame based on their data types and number of unique values.

    Args:
        dataframe (pandas.DataFrame): The input DataFrame.
        cat_th (int, optional): The threshold for considering a column categorical based on the number of unique values. Defaults to 10.
        car_th (int, optional): The threshold for considering a categorical column with many unique values as cardinal. Defaults to 20.

    Returns:
        tuple: A tuple containing three lists:
            - cat_cols (list): List of column names identified as categorical.
            - num_cols (list): List of column names identified as numerical.
            - cat_but_car (list): List of column names initially identified as categorical but considered cardinal due to exceeding the car_th threshold.
    """

    # Step 1: Identify initial categorical columns (object type)
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    print(f"# Initial Categorical Columns (object type): {len(cat_cols)}")

    # Step 2: Identify numerical-looking categorical columns (low unique values, not object type)
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                    dataframe[col].dtypes != "O"]
    print(f"# Categorical-looking Numerical Columns (< {cat_th} unique values, not object type): {len(num_but_cat)}")

    # Step 3: Identify and remove cardinal columns (high unique values, object type) from initial categorical list
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                    dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat  # Combine initial and numerical-looking categorical
    cat_cols = [col for col in cat_cols if col not in cat_but_car]  # Remove cardinal columns
    print(f"# Cardinal Columns (> {car_th} unique values, object type): {len(cat_but_car)}")

    # Step 4: Identify final numerical columns (not object type, excluding previously identified)
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]  # Remove numerical-looking categorical
    print(f"# Final Numerical Columns (not object type, excluding identified categorical): {len(num_cols)}")

    # Step 5: Report summary of column categories
    print(f"\nObservations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')  # the number of categorical variables
    print(f'num_cols: {len(num_cols)}')  # the number of numerical variables
    print(f'cat_but_car: {len(cat_but_car)}')  # the number of cardinal variables
    print(f'num_but_cat: {len(num_but_cat)}')  # the number of categorical variables that look numerical

    return cat_cols, num_cols, cat_but_car
```

"Cardinal" refers to a type of **variable that falls somewhere between numerical and categorical.** It has characteristics of both:

*   Like categorical variables:
    * It typically holds non-numerical text-like data (e.g., strings).
    * It doesn't naturally have a numerical order or meaningful mathematical operations.
*   Like numerical variables:
    * It has a high number of unique values, often approaching the number of observations.

The code identifies cardinal variables based on two criteria:

1.  Data Type: It considers columns with the "object" data type as potential candidates, as this is a common type for text-like data.
*   Number of Unique Values: It sets a threshold (car_th in the code, defaulting to 20). If a column has more unique values than this threshold, it's classified as cardinal.

Examples of cardinal variables:

*   Product IDs.
*   Customer IDs.
*   Addresses.
*   Long text descriptions.

Key points for cardinal variables:

*   They aren't inherently numerical or categorical.
*   They require careful handling in data analysis, as techniques for categorical variables might not always apply directly, and using them as numerical could lead to issues.
*   Domain knowledge about the data is crucial to appropriately interpret and utilize cardinal variables in analysis.

Imagine a big group of toys, all different shapes and sizes! 🪁

This code is like a special tool that helps us find the toys that stand out from the crowd!

It works like this:

We line up all the toys from smallest to biggest.
We find the toys that mark the middle of the line. ➡️➡️
We draw a special fence around most of the toys, leaving out only a few at each end.
The toys that are outside the fence are the special ones we call "outliers"!
It's like finding the hidden treasures in the toy box! ✨

```python
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
for i in df.columns:
    print("Thresholds of {} : ({:.2f}, {:.2f})".format(i, *outlier_thresholds(df,i)))
```

This function takes three arguments: a dataframe, the name of a column in the dataframe, and two optional numbers (quartiles). The function calculates the quartiles of the data in the specified column. The quartiles are the values that divide the data into four equal parts. The function then calculates the interquartile range (IQR), which is the difference between the third quartile and the first quartile. The IQR is a measure of how spread out the data is. Finally, the function calculates the upper and lower outlier thresholds. These are the values that are 1.5 times the IQR below the first quartile and 1.5 times the IQR above the third quartile.

After the outlier thresholds are calculated, the code would then filter the dataframe to only include the rows where the value in the "Pregnancies" column is less than the lower outlier threshold or greater than the upper outlier threshold. This would create a new dataframe that only contains the outliers.

In other words, this code would find the data points in the "Pregnancies" column that are far away from the rest of the data and then create a new dataframe that only contains those data points.

1. df[ ... ]: This part selects rows from the DataFrame df based on a condition within the square brackets.

2. (df["Pregnancies"] < low): This checks if the value in the "Pregnancies" column for each row is less than the value stored in the variable low.
   * Imagine comparing each pregnancy count in the data to the low threshold.

3. |: This symbol represents the logical OR operator.
   * It means the condition can be true either if the previous condition (< low) is met OR if the next condition (> up) is met.

4. (df["Pregnancies"] > up): This checks if the value in the "Pregnancies" column for each row is greater than the value stored in the variable up.

### Missing Values

Missing values, also known as null values, can be a headache when working with data. They're like empty spaces in your information, and how you handle them depends on the situation. Here's a breakdown of what you can do:

1. Identify the Missing Values:

  The first step is to figure out where the missing data is. You can use functions like `df.isnull().sum()` to see how many missing values each column has. This helps you understand the extent of the problem.

2. Understand Why They're Missing:

  If possible, try to understand why there's missing data. Did people forget to fill out a form? Was there a technical issue? Knowing the reason can help you decide the best way to handle them.

3. Remove the Rows/Columns (Deletion):

  This is a simple approach, but it can be risky. If there are only a few missing values and the rest of the data is good, you can remove the rows or columns with missing data entirely. However, this throws away information, so it's not ideal if you have a lot of missing data or it's not random.

4. Fill the Missing Values (Imputation):

  * This approach involves estimating what the missing values might be and filling them in with those estimates. There are different ways to do this:
    * Fill with a constant: You can replace missing values with a constant value, like the average of the column. This is okay for numerical data but not ideal for categorical data (e.g., filling hair color with "brown" wouldn't be accurate).
    * Use statistical methods: You can use methods like mean, median, or mode to estimate the missing values based on the surrounding data. This works better with numerical data and assumes the missing values are random.
    * Use machine learning: For more complex datasets, you can use machine learning models to predict the missing values based on the other data points in the row. This can be more accurate but requires more expertise.

5. Consider Leaving Them As Is:

  In some cases, leaving the missing values as they are might be the best option. This could be the case if there's a good reason for the missing data (e.g., someone didn't answer a question that didn't apply to them) or if you don't have enough information to fill them accurately.

The best approach depends on your specific data and analysis goals. Consider the amount of missing data, why it's missing, and the type of data (numerical vs categorical) when making your decision.

This function, named missing_values_table, takes a pandas DataFrame (dataframe) as input and optionally a boolean value (na_name) as an argument. It analyzes the DataFrame and generates a table summarizing missing values.

  This line calculates the number of missing values `(isnull().sum())` in each column listed in `na_columns`. It then sorts the resulting Series `(n_miss)` in descending order using `sort_values(ascending=False)`. This prioritizes displaying columns with higher numbers of missing values first.

4. Printing the Table and Returning Column Names:

  `print(missing_df, end="\n")`

  This line simply prints the constructed `missing_df` to the console, displaying a clear table summarizing missing values. The `end="\n"` argument ensures a new line is printed after the table.
  
  `if na_name:`

  This conditional statement checks if the optional argument `na_name` is set to `True`.

```python
def missing_values_table(dataframe, na_name=False):
    # only take missing columns
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]

    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)  # number of missing value
    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False) # ratio
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])  # make table
    print(missing_df, end="\n")

    if na_name:
        return na_columns
```

This code investigates whether missing values in specific columns have any connection to a particular target variable (in this case, "Outcome"). It's a common practice in data cleaning and exploration to understand how missing data might impact analysis.

2. Copy Data:

3. Flag Missing Values:

  * For each column in na_columns:
    * The function creates a new column with the name column_name_NA_FLAG.

* This information can guide further data cleaning decisions and potential handling of missing values.



```python
def missing_vs_target(dataframe, target, na_columns):
    temp_df = dataframe.copy()

    for col in na_columns:
        temp_df[col + '_NA_FLAG'] = np.where(temp_df[col].isnull(), 1, 0)

    na_flags = temp_df.loc[:, temp_df.columns.str.contains("_NA_")].columns

    for col in na_flags:
        print(pd.DataFrame({"TARGET_MEAN": temp_df.groupby(col)[target].mean(),
                            "Count": temp_df.groupby(col)[target].count()}), end="\n\n\n")

    print("*********** Changed DataFrame ***********")
    print(temp_df.head(10))


missing_vs_target(df, "Outcome", na_cols)
```

## Solve Outliers and Missing Values

There are several approaches to handling missing values in a dataset. The best solution depends on the specific situation, including:

* The amount of missing data
* The reason for the missing data (Missing Completely At Random (MCAR), Missing At Random (MAR), Missing Not At Random (MNAR))
* The type of data (numerical or categorical)

#### Here are some common techniques for dealing with missing values:

*   List-wise Deletion: This involves removing entire rows or columns that contain missing values. This can be a quick solution, but it can also lead to a loss of data, especially if the missing data percentage is high.
*   Mean/Median/Mode Imputation: This involves replacing missing values with the average (mean), middle value (median), or most frequent value (mode) for the respective column. This is a simple approach, but it can introduce bias, especially for skewed data (mean) or if the missing values are not random (median/mode).
*   Regression Imputation: This method uses a statistical model (like linear regression) to predict the missing value based on other variables in the dataset. This can be a more sophisticated approach, but it requires careful model selection and assumptions about the data.
*   MICE (Multiple Imputation by Chained Equations): This technique creates multiple complete versions of the dataset by iteratively imputing missing values based on other variables. This helps account for the uncertainty associated with missing data.
*   Using Missingness as a Feature: In some cases, you can create a new feature to indicate whether a value is missing or not. This can be useful for some machine learning models.

#### Choosing the Right Approach:

The best way to handle missing values depends on the specifics of your data and analysis. Here are some general guidelines:

* If the missing data percentage is low, and the missingness is random (MCAR), list-wise deletion or simple imputation might be sufficient.
* If the missing data percentage is higher or the missingness is not random (MAR or MNAR), consider using more sophisticated techniques like regression imputation or MICE.
* Always document the approach you take to handle missing values, as it can impact the results of your analysis.

## Outliers

**Local Outlier Factor (LOF)** is an unsupervised anomaly detection algorithm that identifies outliers based on the local density deviation of a data point compared to its neighbors. It's a useful technique for finding anomalies within clusters of data points.

Here's how LOF works:

1. K-Nearest Neighbors (KNN): LOF relies on the concept of K-Nearest Neighbors (KNN). For each data point, it identifies its k nearest neighbors in the dataset.

2. Reachability Distance (RD): It calculates the reachability distance for each neighbor. The reachability distance is essentially the "distance" to reach a neighbor, considering the neighbor's own density. It's not just the raw Euclidean distance.

3. Local Reachability Density (LRD): Based on the reachability distances of all its neighbors, LOF calculates the Local Reachability Density (LRD) for a data point. This LRD represents the average reachability distance of its neighbors.

4. Local Outlier Factor (LOF): Finally, LOF calculates the actual Local Outlier Factor for the data point. This is the ratio between the LRD of the data point and the LRD of its neighbors.

* LOF > 1: If a data point has a significantly lower density compared to its neighbors (higher LRD), its LOF will be greater than 1. This suggests it's a potential outlier.
* LOF ≈ 1: Points with LOF close to 1 have similar density to their neighbors and are likely not outliers.
* LOF < 1: In rare cases, points might have a higher density than their neighbors (lower LRD), resulting in an LOF less than 1. These might be interesting data points to investigate further.

In our dataset, there are both "**Age**" and "**Pregnancies**". For instance, when these variables are taken separately, there may not be any outlier. However, when these 2 variables get together, it may be the outlier.

If a variable is converted like this to numeric values and sent to algorithms, the algorithm defines this variable as a variable ranging from an interval. In other words, this transformation will confuse the algorithms. The reason for its corruption is here, the categorical variable has turned into numerical values, and a categorical variable, which is normally a nominal scale type, is suddenly distorted by a transformation process and is subjected to a sorting process as if there is a difference between them. The characteristic of the nominal scale type is that the distance between classes is equal, but this equality is disturbed after this transformation.

This situation can cause biases and problems when we convert categorical variables to numeric values and confuse algorithms. In this case, we need to do **one-hot encoding**.

To fix this problem, one-hot conversion is done, but doing this also creates a trap named as "dummy variable trap."

If we apply a transformation to the variables in the data set, if the new variables can be created over each other, this is a dummy variable trap. In other words, if there is a variable that represents another variable, this situation is called a dummy variable trap.

```python
df_new = pd.get_dummies(df_apply_median_ml, columns=["Patient"])
df_new.head()
```

As it can be seen, Patient is dropped, and there are new variables: Patient_No and Patient_Yes. In that situation, we have to drop one of these, and it helps to avoid creating over each other and duplication. The dummy variable trap is solved with this.

```python
df_new = pd.get_dummies(df_apply_median_ml, columns=["Patient"], drop_first=True)
df_new.head()
```

### Scaling

The aim is to eliminate the measurement difference between the variables and to try to ensure that the models to be used approach the variables under equal conditions. Tree-based methods are not affected by outlier, standardization. In the general trend, we may prefer to scale features.

The variance structure and information structure within the variable itself are not deteriorated, but are set to a certain standard. For example, let's say a dataset has a value of 10 and is ranked 80th when ordered from smallest to largest. When this variable is standardized, the value of 10 will be something like between 1-2 or 0 to 1, but when the data set is again ordered from smallest to largest, this value will still correspond to the 80th value. Therefore, when a variable is standardized, the value of the variable will change, it will be put into a certain format, but the spread and the essence of the distribution information will not change. The values of the variable such as mean, standard deviation and variance will change,as well, but the spread, distribution and current state of the information it carries within the variable will not change.

There are 3 common scaling method:

1. Standard Scaler: Normalization. Subtract the mean, divide by the standard deviation. It can be called z-standardization. z = (x - u) / s.

  The Standard Scaler subtracts the mean from all observation units and divides it by the standard deviation.
  CAUTION: Both standard deviation and mean are metrics that are affected by outliers. Therefore, if we subtract the median from all observation units and divide by the IQR, which is not affected by outliers, we will both consider the central tendency and change, and perform a more robust standardization process. Robust Scaler is more preferable because it is resistant to outliers compared to Standard Scaler, but its use is not common. Usually Standard Scaler and MinMax are used.

2. Robust Scaler: Subtract median and divide by IQR.

3. MinMaxScaler: It is used to convert values between 2 desired ranges. These 2 values may also be values that are not in our dataset.

```python
df_scale = df_cat_modified.copy()

# standart scaler
ss = StandardScaler()
df_scale["Age_standard_scaler"] = ss.fit_transform(df_scale[["Age"]])
df_scale.head()

```

```python
# min-max scaler
# The range can be given with the feature_range=() argument
mms = MinMaxScaler() # default range from 0 to 1
df_scale["Age_min_max_scaler"] = mms.fit_transform(df_scale[["Age"]])

df_scale.head()
```

The difference between scales can be observed above. In this problem, "robust scaler" will be used here. There is no specific reason, simply, all scaling types aim for the same thing, but impact in different ways.
